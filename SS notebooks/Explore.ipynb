```python
# plot a correlation heat map of selected columns in the dataset
selected_cols = ["temp", "area", "rain", "wind"]

mask = np.triu(np.ones_like(df[selected_cols].corr(), dtype=bool))

sns.heatmap(df[selected_cols].corr(), annot=True, mask=mask)
plt.show()
```

```python
# plot a correlation heat map of selected columns in the dataset
selected_cols = ["area", "FFMC", "DMC", "ISI"]

mask = np.triu(np.ones_like(df[selected_cols].corr(), dtype=bool))

sns.heatmap(df[selected_cols].corr(), annot=True, mask=mask)
plt.show()
```
